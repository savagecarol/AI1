```python
import numpy as np
```

```python
x = np.arange(0 , 10)
```

```python
y = np.arange(11 , 21)
```

```python
plt.scatter(x , y , c = 'g')
plt.xlabel('X label')
plt.ylabel('y label')
plt.title('Title')
plt.savefig('first.png')
```
